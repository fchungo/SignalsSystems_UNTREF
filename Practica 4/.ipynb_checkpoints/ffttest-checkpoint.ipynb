```python
from scipy.fftpack import fft, ifft, fftshift
import scipy.io.wavfile as spwav
import numpy as np
import time

def calculo_fft(audio_data):
    inicio = time.time()
    espectro_audio = fft(audio_data)
    espectro_audio_norm = np.abs(espectro_audio)
    espectro_audio_inv = ifft(espectro_audio)
    return espectro_audio_norm, espectro_audio_inv

def cargar_wav(archivo):
    audiodata = spwav.read(archivo)
    return audiodata
```

```python
#archivo = '/home/fabri/Escritorio/rec_20190318175117.wav'
archivo1 = '/home/fabri/Escritorio/TESIS/programacion/csonora_python/csonora/lib/audio/IR_dispositivo_mar19.wav'
#archivo = '/home/fabri/Escritorio/TESIS/Grabaciones/prueba_audio_usb.wav'
archivo2 = '/home/fabri/Escritorio/TESIS/programacion/csonora_python/csonora/lib/audio/IR_referencia_mar19.wav'
#archivo = '/home/fabri/Escritorio/TESIS/programacion/csonora_python/csonora/lib/audio/calibracion.wav'

fs, audiosig_disp = cargar_wav(archivo1)
fs, audiosig_ref = cargar_wav(archivo2)
```

```python
audiofft_disp, inv_audiofft_disp = calculo_fft(audiosig_disp)
audiofft_ref,inv_audiofft_ref = calculo_fft(audiosig_ref)
```
